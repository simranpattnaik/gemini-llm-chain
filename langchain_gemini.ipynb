```python
# ================= INSTALL (RUN ONCE) =================
import sys
!{sys.executable} -m pip install -q langchain langchain-experimental langchain-google-genai pandas

# ================= IMPORTS =================
import os
import pandas as pd
from langchain_experimental.agents import create_pandas_dataframe_agent
from langchain_google_genai import ChatGoogleGenerativeAI

# ================= SET GEMINI API KEY =================
# 🔑 Get from: https://aistudio.google.com/app/apikey
os.environ["GOOGLE_API_KEY"] = "AIzaSyAHH12aN-RQb0ofKNI7st8kB7K0p4RO3zU"

# ================= LOAD DATA =================
df = pd.read_csv("titanic dataset.csv")

# Cleaned dataframe
df1 = df.copy()
df1["Age"] = df1["Age"].fillna(df1["Age"].mean())

# ================= GEMINI MODEL =================
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash-lite",   # fast + free tier friendly
    temperature=0
)

# ================= CREATE AGENT =================
agent = create_pandas_dataframe_agent(
    llm,
    [df, df1],
    verbose=True,
    allow_dangerous_code=True
)

# ================= QUICK TEST =================
print("\n🔍 Rows in dataset:")
print(agent.invoke("How many rows are there in df?"))

print("\n🔍 Average Age:")
print(agent.invoke("What is the average age in df1?"))

print("\n🔍 Missing Values Comparison:")
print(agent.invoke("Compare missing values in df and df1"))

# ================= INTERACTIVE CHAT =================
while True:
    q = input("\n💬 Ask about your data (type 'exit'): ")
    if q.lower() == "exit":
        print("👋 Done.")
        break
    
    try:
        response = agent.invoke(q)
        print("\n🤖:", response)
    except Exception as e:
        print("⚠️ Error:", e)
```
